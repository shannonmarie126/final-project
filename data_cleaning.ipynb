```python
income_bins = [0, 25000, 50000, 75000, 100000, 125000, 150000, 175000, 200000, 225000]
```

```python
df['income_ranges'] = pd.cut(df["Income"], income_bins)

df.head()
```

```python
df["income_ranges"].value_counts()
```

```python
age_bins = [17, 39, 59, 81]
labels = ["young adult", "adult", "senior"]

df['Age_Ranges'] = pd.cut(df["Age"], bins=age_bins, labels=labels)
df.head()
```

```python
df["Age_Ranges"].value_counts()
```
